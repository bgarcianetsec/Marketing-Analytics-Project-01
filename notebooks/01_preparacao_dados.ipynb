```python
# Usando o 'r' para evitar problemas com as barras do Windows
caminho_arquivo = r'C:\ProjetosDados\data\marketing_campaign.csv'

# Lendo o arquivo com o separador correto
df = pd.read_csv(caminho_arquivo, sep=';')

# Verificando se agora as colunas estão separadas
print("--- Primeiras Linhas do DataFrame ---")
display(df.head())

# Verificando os valores ausentes novamente
print("\n--- Valores Ausentes por Coluna ---")
print(df.isnull().sum())
```

```python
# Preenchendo valores nulos na Renda com a mediana (para evitar distorção por valores muito altos)
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
df.isnull().sum()
```

```python
#Verificando idades extremas e rendas
print(df[['Year_Birth', 'Income']].describe())

#Filtrando para ver se alguém nasceu antes de 1940 (Exemplo)
print("\n--- Clientes com ano de nascimento suspeito")
print(df[df['Year_Birth'] < 1940])
```

```python
#Filtrando o DataFrame para remover valores abusurdos
df = df[df['Year_Birth'] > 1920]
df = df[df['Income'] < 200000]

print(f"Novo tamanho do dataset: {df.shape[0]} linhas")
```

```python
# Criando a coluna Idade baseada no ano atual (2026)
df['Age'] = 2026 - df['Year_Birth']

# Verificando a nova distribuição de idade
print("\n--- Resumo da Idade ---")
print(df['Age'].describe())
```

```python
#Verificando os tipos de status que existem
print("\nStatus Civil Original: ", df['Marital_Status'].unique())

```

```python
# Agrupando em categorias mais simples
df['Marital_Status'] = df['Marital_Status'].replace({
    'Married': 'Casado',
    'Together': 'Casado',
    'Absurd': 'Solteiro',
    'Widow': 'Solteiro',
    'YOLO': 'Solteiro',
    'Divorced': 'Solteiro',
    'Alone': 'Solteiro'
})
```

```python
#Vericando alteraçõe da coluna estado civil 
print("\nStatus Civil reorganizado: ", df['Marital_Status'].unique())
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Single': 'Solteiro'
})
```

```python
print(df.info())
```

```python
#Removendo colunas
df.drop(columns=['Z_CostContact', 'Z_Revenue'], inplace=True, errors='ignores')
```

```python
# 1. ENGENHARIA DE ATRIBUTOS (Criando as colunas que faltam)
# Somando todas as colunas de gastos em uma única coluna 'Total_Mnt'
colunas_gastos = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Mnt'] = df[colunas_gastos].sum(axis=1)

# Criando a coluna de idade (ajuste o ano se necessário)
df['Age'] = 2026 - df['Year_Birth']

# 2. AGORA SIM, OS GRÁFICOS
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 5]

# --- PERGUNTA 1: Escolaridade vs Gasto Total ---
plt.figure()
ordem_edu = ['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']
sns.barplot(data=df, x='Education', y='Total_Mnt', order=ordem_edu, palette='viridis', estimator='mean', errorbar=None)
plt.title('Média de Gasto Total por Nível de Escolaridade')
plt.show()

# --- PERGUNTA 2: Canais de Venda ---
plt.figure()
canais = {
    'Loja': df['NumStorePurchases'].sum(),
    'Web': df['NumWebPurchases'].sum(),
    'Catálogo': df['NumCatalogPurchases'].sum()
}
plt.bar(canais.keys(), canais.values(), color=['#4C72B0', '#55A868', '#C44E52'])
plt.title('Onde os clientes mais compram? (Volume Total)')
plt.show()

# --- PERGUNTA 3: Perfil de Renda de quem aceitou a Campanha ---
plt.figure()
sns.boxplot(data=df, x='Response', y='Income', palette='Set2')
plt.title('Distribuição de Renda: Recusou (0) vs Aceitou (1)')
plt.ylim(0, 120000) 
plt.show()
```
